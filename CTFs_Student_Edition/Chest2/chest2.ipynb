```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wav
from IPython.display import Audio
```

find twins

```python
# Load WAV files
filenames = ['output3.wav', 'output4.wav', 'output5.wav']
audio_data = {f: wav.read(f) for f in filenames}

# Plot waveforms
plt.figure(figsize=(12, 8))
for i, fname in enumerate(filenames):
    rate, data = audio_data[fname]
    time = np.linspace(0, len(data) / rate, num=len(data))
    plt.subplot(3, 1, i + 1)
    plt.plot(time, data)
    plt.title(f"Waveform of {fname}")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
plt.tight_layout()
plt.show()
```

### from the waveforms, the **twins are output3 and aoutput5**

```python
# Load WAV files
rate3, audio3 = wav.read("output3.wav")
rate4, audio4 = wav.read("output4.wav")

# trim to same length
min_len = min(len(audio3), len(audio4))
audio3 = audio3[:min_len]
audio4 = audio4[:min_len]

# XOR to recover hidden message
decoded_audio = np.bitwise_xor(audio3, audio4)

# Save the shuffled but audible audio for manual segmenting
shuffled_audio_path = "./shuffled_decoded.wav"
wav.write(shuffled_audio_path, rate3, decoded_audio)

# Try to listen
Audio(decoded_audio, rate=rate4)
```
